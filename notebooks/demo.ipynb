# Agentic AI RAG Demo Walkthrough

This notebook demonstrates how to use the core components of the project:

- `Embedder` for embeddings
- `VectorStore` for FAISS-based similarity search
- `Retriever` to fetch relevant documents
- `Agent` to orchestrate retrieval + summarization


```python
from rag.embedder import Embedder
from rag.vector_store import VectorStore
from rag.retriever import Retriever
from agent.agent_core import Agent

embedder = Embedder()
store = VectorStore("../data/sample_docs")
retriever = Retriever(embedder, store)
agent = Agent(retriever=retriever)

query = "Summarize the documents."
response = agent.run(query)
response
```
